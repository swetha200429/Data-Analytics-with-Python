```python
import seaborn as sns
import matplotlib.pyplot as plt
categories=["Pizza","Burger","Biryani","Pasta","Desser"]
orders=[120,150,180,90,110]
sns.set_palette("coolwarm")
sns.barplot(x=categories,y=orders)
plt.title("Food Category Orders")
plt.xlabel("Category")
plt.ylabel("Number of Orders")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
order_amounts=[120,150,200,220,250,300,320,150,180,210,230,260,280,300,350,400,450,500]
sns.histplot(order_amounts,bins=6,kde=True)
plt.title("Order Amount Distribution")
plt.xlabel("Order Amount")
plt.ylabel("Frequency")
plt.grid()
plt.show()
```
