```python
import numpy as np
temperature_int=np.array([30,56,34,90,78])
print("Integer Temperature:",temperature_int)
print("Data Type:",temperature_int.dtype)
temperature_float = temperature_int.astype('f')
print("\nConverted to Float:", temperature_float)
print("Data Type:", temperature_float.dtype)
high_temp=temperature_float>30
print("\nHigh Temperature Flags:",high_temp)
print("Data Type:",high_temp.dtype)
print("\nMemory Used (int):", temperature_int.nbytes, "bytes")
print("Memory Used (float):", temperature_float.nbytes, "bytes")
```

```python
import numpy as np
book_ids=np.array([101,445,380,263,537,906])
print("Library Book IDs:",book_ids)
search_id=380
result=np.where(book_ids==search_id)
print(f"\nBook ID {search_id} found at position:",result[0])
high_id_books=np.where(book_ids>200)
print("\nBooks with ID>200 found at position:",high_id_books[0])
```
