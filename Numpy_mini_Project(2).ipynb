```python
import numpy as np
sales=np.array([
    [2000,4300,8900,1200],
    [4900,7689,2134,4300],
    [3000,5488,1990,5000],
    [6800,5433,7666,1999]
])
print("Full Sales Data:\n",sales)
q1=sales[0:2]
print("\nFirst 2 months:\n",q1)
q2=sales[-3:]
print("\nLast 3 months:\n",q2)
q3=sales[:,1]
print("\nProduct 2 Sales:\n",q3)
subset=sales[0:3,0:1]
print("\nSubset Data:\n",subset)
```

```python
import numpy as np
scores=np.array([540,745,230,600,894])
print("Original Scores:",scores)
sorted_scores=np.sort(scores)
print("Sorted Scores (Ascending):",sorted_scores)
level_scores=np.array([
    [540,745,230],
    [600,894,900]
])
print("\nOriginal Level Scores:\n",level_scores)
sorted_level_scores=np.sort(level_scores)
print("\nSorted Level Scores (Row-wise):\n",sorted_level_scores)
```
