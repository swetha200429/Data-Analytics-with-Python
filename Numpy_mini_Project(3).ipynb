```python
import numpy as np
weather=np.array([
    [35,78,23,56],
    [65,12,89,45],
    [67,90,36,87]
])
print("Weekly Weather Data:\n",weather)
print("\nCity-wise temperature:")
for city in weather:
    print(city)
print("\nAll temperature values:")
for temp in np.nditer(weather):
    print(temp,end=" ")
print("\n\nTemperature with position:")
for index, value in np.ndenumerate(weather):
    print("City", index[0],"Day",index[1],"=",value)
print("\nHigh Temperature Days(>20):")
for temp in np.nditer(weather):
    if(temp>20):
        print(temp)
```
